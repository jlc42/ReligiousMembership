```python
#Process CCES Data
ccesYears=[]
ccesdfs={}

print('Processing CCES Year 2008')
cces2008df['year']=2008
cces2008df.rename(columns={"V201":"commonweight"}, inplace=True)
cces2008df.rename(columns={"V217":"pew_churatd"}, inplace=True)
cces2008df.rename(columns={"V235":"religpew_mormon"}, inplace=True)
cces2008WeightTotal=cces2008df['commonweight'].sum()
resultsdf.loc[2008,'ccesWeightTotal']=cces2008WeightTotal
resultsdf.loc[2008,'ccesN']=len(cces2008df)
ccesYears.append(2008)
ccesdfs[2008]=cces2008df

print('Processing CCES Year 2009')
cces2009df['year']=2009
cces2009df.rename(columns={"v200":"commonweight"}, inplace=True)
cces2009df.rename(columns={"v217":"pew_churatd"}, inplace=True)
cces2009df.rename(columns={"v235":"religpew_mormon"}, inplace=True)
cces2009WeightTotal=cces2009df['commonweight'].sum()
resultsdf.loc[2009,'ccesWeightTotal']=cces2009WeightTotal
resultsdf.loc[2009,'ccesN']=len(cces2009df)
ccesYears.append(2009)
ccesdfs[2009]=cces2009df

print('Processing CCES Year 2010')
cces2010df['year']=2010
cces2010df.rename(columns={"V101":"commonweight"}, inplace=True)
cces2010df.rename(columns={"V217":"pew_churatd"}, inplace=True)
cces2010df.rename(columns={"V235":"religpew_mormon"}, inplace=True)
cces2010WeightTotal=cces2010df['commonweight'].sum()
resultsdf.loc[2010,'ccesWeightTotal']=cces2010WeightTotal
resultsdf.loc[2010,'ccesN']=len(cces2010df)
ccesYears.append(2010)
ccesdfs[2010]=cces2010df

print('Processing CCES Year 2011')
cces2011df['year']=2011
cces2011df.rename(columns={"V101":"commonweight"}, inplace=True)
cces2011df.rename(columns={"V217":"pew_churatd"}, inplace=True)
cces2011df.rename(columns={"V235":"religpew_mormon"}, inplace=True)
cces2011WeightTotal=cces2011df['commonweight'].sum()
resultsdf.loc[2011,'ccesWeightTotal']=cces2011WeightTotal
resultsdf.loc[2011,'ccesN']=len(cces2011df)
ccesYears.append(2011)
ccesdfs[2011]=cces2011df

print('Processing CCES Year 2012')
cces2012df['year']=2012
cces2012df.rename(columns={"weight_vv":"commonweight"}, inplace=True)
cces2012WeightTotal=cces2012df['commonweight'].sum()
resultsdf.loc[2012,'ccesWeightTotal']=cces2012WeightTotal
resultsdf.loc[2012,'ccesN']=len(cces2012df)
ccesYears.append(2012)
ccesdfs[2012]=cces2012df

print('Processing CCES Year 2013')
cces2013df['year']=2013
cces2013df.rename(columns={"weight":"commonweight"}, inplace=True)
cces2013WeightTotal=cces2013df['commonweight'].sum()
resultsdf.loc[2013,'ccesWeightTotal']=cces2013WeightTotal
resultsdf.loc[2013,'ccesN']=len(cces2013df)
ccesYears.append(2013)
ccesdfs[2013]=cces2013df

print('Processing CCES Year 2014')
cces2014df['year']=2014
cces2014df.rename(columns={"weight":"commonweight"}, inplace=True)
cces2014WeightTotal=cces2014df['commonweight'].sum()
resultsdf.loc[2014,'ccesWeightTotal']=cces2014WeightTotal
resultsdf.loc[2014,'ccesN']=len(cces2014df)
ccesYears.append(2014)
ccesdfs[2014]=cces2014df

print('Processing CCES Year 2015')
cces2015df['year']=2015
cces2015df.rename(columns={"weight":"commonweight"}, inplace=True)
cces2015WeightTotal=cces2015df['commonweight'].sum()
resultsdf.loc[2015,'ccesWeightTotal']=cces2015WeightTotal
resultsdf.loc[2015,'ccesN']=len(cces2015df)
ccesYears.append(2015)
ccesdfs[2015]=cces2015df

print('Processing CCES Year 2016')
cces2016df['year']=2016
cces2016WeightTotal=cces2016df['commonweight'].sum()
resultsdf.loc[2016,'ccesWeightTotal']=cces2016WeightTotal
resultsdf.loc[2016,'ccesN']=len(cces2016df)
ccesYears.append(2016)
ccesdfs[2016]=cces2016df

print('Processing CCES Year 2017')
cces2017df['year']=2017
cces2017df.rename(columns={"weights_common":"commonweight"}, inplace=True)
cces2017WeightTotal=cces2017df['commonweight'].sum()
resultsdf.loc[2017,'ccesWeightTotal']=cces2017WeightTotal
resultsdf.loc[2017,'ccesN']=len(cces2017df)
ccesYears.append(2017)
ccesdfs[2017]=cces2017df

print('Processing CCES Year 2018')
cces2018df['year']=2018
cces2018WeightTotal=cces2018df['commonweight'].sum()
resultsdf.loc[2018,'ccesWeightTotal']=cces2018WeightTotal
resultsdf.loc[2018,'ccesN']=len(cces2018df)
ccesYears.append(2018)
ccesdfs[2018]=cces2018df

print('Processing CCES Year 2019')
cces2019df['year']=2019
cces2019WeightTotal=cces2019df['commonweight'].sum()
resultsdf.loc[2019,'ccesWeightTotal']=cces2019WeightTotal
resultsdf.loc[2019,'ccesN']=len(cces2019df)
ccesYears.append(2019)
ccesdfs[2019]=cces2019df

print('Processing CCES Year 2020')
cces2020df['year']=2020
cces2020WeightTotal=cces2020df['commonweight'].sum()
resultsdf.loc[2020,'ccesWeightTotal']=cces2020WeightTotal
resultsdf.loc[2020,'ccesN']=len(cces2020df)
ccesYears.append(2020)
ccesdfs[2020]=cces2020df

resultsdf.sort_index(inplace=True)
```

```python
#Do All Years
yearsToAnalize=ccesYears
#Do Just Some Years
#yearsToAnalize=[2018]

for y in yearsToAnalize:
    print("Analyzing CCES for year " + str(y) + ".")
    df=ccesdfs[y]
    LDSfilter = (df['religpew_mormon']=='The Church of Jesus Christ of Latter-day Saints') | (df['religpew_mormon']=='the church of jesus christ of latter-day saints')
    LDSdf=df.loc[LDSfilter]
    resultsdf.loc[y,'ccesLDSWeightTotal']=LDSdf['commonweight'].sum()
    resultsdf.loc[y,'ccesLDSN']=len(LDSdf)
resultsdf['ccesLDSWeightedFrac'] = resultsdf['ccesLDSWeightTotal']/resultsdf['ccesWeightTotal']
resultsdf['ccesLDSNFrac'] = resultsdf['ccesLDSN']/resultsdf['ccesN']
```
